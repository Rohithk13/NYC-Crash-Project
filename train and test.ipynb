```python
# to fill null values
import random

# List of specific columns targeted for null replacement
target_columns = ["BOROUGH", "ZIP CODE", "LATITUDE", "LONGITUDE", "ON STREET NAME", "CROSS STREET NAME"]

# Isolating distinct, non-null value sets from the DataFrame for the target columns
distinct_non_null_sets = new_pds[target_columns].dropna().drop_duplicates()

# Function to select a random non-null set from the unique combinations
def choose_random_set():
    return distinct_non_null_sets.sample(n=1)

# Applying random non-null sets to fill in nulls across specified columns
for each_column in target_columns:
    chosen_set = choose_random_set()
    new_pds[each_column] = new_pds[each_column].fillna(chosen_set[each_column].values[0])
```

```python
# Check null values in each column
null_values_per_column = new_pds.isnull().sum()

print("Null values in each column:")
print(null_values_per_column)
```

```python
# Function to exclude "Unspecified" values and compute the mode for a given column
def compute_mode(dataframe, column_name):
    mode_value = dataframe[dataframe[column_name] != "Unspecified"][column_name].mode().iloc[0]
    return mode_value

# Compute mode for CONTRIBUTING FACTOR VEHICLE 1 and CONTRIBUTING FACTOR VEHICLE 2
factor_1_mode = compute_mode(new_pds, "CONTRIBUTING FACTOR VEHICLE 1")
factor_2_mode = compute_mode(new_pds, "CONTRIBUTING FACTOR VEHICLE 2")

# Calculate the mode for vehicle type codes without excluding any specific value
vehicle_1_mode = new_pds["VEHICLE TYPE CODE 1"].mode().iloc[0]
vehicle_2_mode = new_pds["VEHICLE TYPE CODE 2"].mode().iloc[0]

# Fill null values with mode values
new_pds["CONTRIBUTING FACTOR VEHICLE 1"].fillna(factor_1_mode, inplace=True)
new_pds["CONTRIBUTING FACTOR VEHICLE 2"].fillna(factor_2_mode, inplace=True)
new_pds["VEHICLE TYPE CODE 1"].fillna(vehicle_1_mode, inplace=True)
new_pds["VEHICLE TYPE CODE 2"].fillna(vehicle_2_mode, inplace=True)

# Filter rows with "Unspecified" in CONTRIBUTING FACTOR VEHICLE 1 and CONTRIBUTING FACTOR VEHICLE 2
new_pds = new_pds[(new_pds["CONTRIBUTING FACTOR VEHICLE 1"] != "Unspecified") & (new_pds["CONTRIBUTING FACTOR VEHICLE 2"] != "Unspecified")]
```

```python
# Convert the 'CRASH TIME' column to strings
new_pds['CRASH TIME'] = new_pds['CRASH TIME'].astype(str)

# Function to categorize time of day with error handling
def categorize_time_of_day(crash_time):
    try:
        hour = int(crash_time.split(':')[0])
        if 5 <= hour < 12:
            return 'Morning'
        elif 12 <= hour < 17:
            return 'Afternoon'
        elif 17 <= hour < 21:
            return 'Evening'
        else:
            return 'Night'
    except ValueError:
        return 'Unknown'

# Apply the function to create a new column 'TIME OF DAY'
new_pds['TIME OF DAY'] = new_pds['CRASH TIME'].apply(categorize_time_of_day)

# Selecting relevant columns for the analysis
classification_df = new_pds[['TIME OF DAY', 'LATITUDE', 'LONGITUDE', 
                             'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 
                             'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 
                             'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 
                             'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']]

# Display the first few rows of the classification dataframe
classification_df.head()
```

```python
# Convert 'CRASH DATE' to datetime to extract the year
new_pds['CRASH DATE'] = pd.to_datetime(new_pds['CRASH DATE'], format='%m/%d/%Y')
new_pds['YEAR'] = new_pds['CRASH DATE'].dt.year

# Determine the range of years in the dataset
years_range = new_pds['YEAR'].unique()
years_range.sort()

# Get the training and testing year ranges
training_years = years_range[:10]
testing_years = years_range[-3:]

# Split the data into training and testing sets
train_df = new_pds[new_pds['YEAR'].isin(training_years)]
test_df = new_pds[new_pds['YEAR'].isin(testing_years)]

train_df.shape, test_df.shape
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import joblib

# Define the features and target variables
features = ['TIME OF DAY', 'LATITUDE', 'LONGITUDE']
target_columns = [
    'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 
    'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 
    'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 
    'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
]

# Preprocessing pipeline for the features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='median'), ['LATITUDE', 'LONGITUDE']),
        ('cat', OneHotEncoder(), ['TIME OF DAY'])
    ])

# Define the model pipeline with a MultiOutputRegressor
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', MultiOutputRegressor(LinearRegression()))
])

# Separate the features and target variables in the training set
X_train = train_df[features]
y_train = train_df[target_columns]

# Train the model
model.fit(X_train, y_train)

# Save the trained model to a file
model_filename = 'multi_output_trained_model.joblib'
joblib.dump(model, model_filename)
```

```python
import joblib
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd

# Load the trained model from the file
model_filename = 'multi_output_trained_model.joblib'
model = joblib.load(model_filename)

# Separate the features and target variables in the testing set
X_test = test_df[features]
y_test = test_df[target_columns]

# Predict on the testing set
y_pred = model.predict(X_test)

# Evaluate the model for each target variable
evaluation_metrics = {}
for i, column in enumerate(target_columns):
    mae = mean_absolute_error(y_test.iloc[:, i], y_pred[:, i])
    mse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i])
    rmse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i], squared=False)
    evaluation_metrics[column] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse}

# Print evaluation metrics
for column, metrics in evaluation_metrics.items():
    print(f'{column}: MAE = {metrics["MAE"]}, MSE = {metrics["MSE"]}, RMSE = {metrics["RMSE"]}')

# Create a dataframe to display actual and predicted values
results_df = pd.DataFrame(y_test)
predicted_df = pd.DataFrame(y_pred, columns=[f'Predicted {col}' for col in target_columns])
results_df = pd.concat([results_df.reset_index(drop=True), predicted_df.reset_index(drop=True)], axis=1)

# Display the first few rows of the results dataframe
results_df.tail(10)
```

```python
# Visualize the actual vs predicted values for each target variable
plt.figure(figsize=(20, 20))
for i, column in enumerate(target_columns):
    plt.subplot(4, 2, i + 1)
    plt.scatter(results_df[column], results_df[f'Predicted {column}'], alpha=0.5)
    plt.plot([results_df[column].min(), results_df[column].max()],
             [results_df[column].min(), results_df[column].max()],
             'r--', lw=2)
    plt.xlabel(f'Actual {column}')
    plt.ylabel(f'Predicted {column}')
    plt.title(f'Actual vs Predicted {column}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import joblib


# Define the model pipelines with MultiOutputRegressor
rf_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42)))
])

 # Separate the features and target variables in the training set
X_train = train_df[features]
y_train = train_df[target_columns]

# Train the models
rf_model.fit(X_train, y_train)

# Save the trained models to files
joblib.dump(rf_model, 'rf_model.joblib')
```

```python
import joblib
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt

# Load the trained Random Forest model from the file
rf_model = joblib.load('rf_model.joblib')

# Separate the features and target variables in the testing set
X_test = test_df[features]
y_test = test_df[target_columns]

# Predict on the testing set
y_pred_rf = rf_model.predict(X_test)

# Evaluate the Random Forest model for each target variable
evaluation_metrics_rf = {}
for i, column in enumerate(target_columns):
    mae_rf = mean_absolute_error(y_test.iloc[:, i], y_pred_rf[:, i])
    mse_rf = mean_squared_error(y_test.iloc[:, i], y_pred_rf[:, i])
    rmse_rf = mean_squared_error(y_test.iloc[:, i], y_pred_rf[:, i], squared=False)
    evaluation_metrics_rf[column] = {'MAE': mae_rf, 'MSE': mse_rf, 'RMSE': rmse_rf}

# Print evaluation metrics for Random Forest model
print("Random Forest Evaluation Metrics:")
for column, metrics in evaluation_metrics_rf.items():
    print(f'{column}: MAE = {metrics["MAE"]}, MSE = {metrics["MSE"]}, RMSE = {metrics["RMSE"]}')

# Create a dataframe to display actual and predicted values for Random Forest
results_rf_df = pd.DataFrame(y_test)
predicted_rf_df = pd.DataFrame(y_pred_rf, columns=[f'Predicted {col}' for col in target_columns])
results_rf_df = pd.concat([results_rf_df.reset_index(drop=True), predicted_rf_df.reset_index(drop=True)], axis=1)

results_rf_df.head()

# Visualize the actual vs predicted values for each target variable for Random Forest
plt.figure(figsize=(20, 20))
for i, column in enumerate(target_columns):
    plt.subplot(4, 2, i + 1)
    plt.scatter(results_rf_df[column], results_rf_df[f'Predicted {column}'], alpha=0.5)
    plt.plot([results_rf_df[column].min(), results_rf_df[column].max()],
             [results_rf_df[column].min(), results_rf_df[column].max()],
             'r--', lw=2)
    plt.xlabel(f'Actual {column}')
    plt.ylabel(f'Predicted {column}')
    plt.title(f'Random Forest: Actual vs Predicted {column}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
